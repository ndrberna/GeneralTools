### Scraping and hashing

```python
def parse(url):
    
    
    headers = {
        'user-agent': random.choice(headers_list)}

    try:
        r  = requests.get(url, headers=headers, timeout=8)
        if r.status_code == 200:
            data = r.text
            soup = BeautifulSoup(data, 'lxml')
            output=str([tag.name for tag in soup.find_all()])
            hash=tag2hash(output)
            print url, hash
        else:
            hash=""
    except:
        hash=""
        pass
    return url,hash

def parallel_domains_hashing(file_name,num_processes,start_time):
    print file_name
    with open(filename) as f:
        lines = f.read().splitlines()
        f.close()
    #print lines
    
    
    with closing(Pool(processes=num_processes)) as p:
        records = p.map(parse, lines)
        p.terminate()
    if len(records) > 0:
        with open('output/'+str(start_time), 'w') as f:
            for i,j in records:
                if j:
                    f.write(i+" , "+str(j.value)+'\n')
        f.close()
    
    
```
